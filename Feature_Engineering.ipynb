```python
train_values = pd.read_csv('/content/drive/MyDrive/datasets_tp2/train_values.csv')
```

```python
df_test = pd.read_csv('/content/drive/MyDrive/datasets_tp2/test_values.csv')
```

```python
train_values.drop(['geo_level_2_id', 'geo_level_3_id'], axis = 1, inplace= True)
```

```python
df_test.drop(['geo_level_2_id', 'geo_level_3_id'], axis = 1, inplace= True)
```

```python
train_values['mud_stone_per_geo_level_1'] = train_values[['geo_level_1_id', 'has_superstructure_mud_mortar_stone']]\
.groupby('geo_level_1_id').transform('sum')
```

```python
train_values
```

```python
train_values['avg_floors_per_year'] = train_values[['age', 'count_floors_pre_eq']].groupby('age').transform('mean')
```

```python
train_values['avg_area'] = train_values[['geo_level_1_id', 'area_percentage']].\
groupby('geo_level_1_id').transform('mean')
```

## Binary Enconding

```python
df_test['mud_stone_per_geo_level_1'] = df_test[['geo_level_1_id', 'has_superstructure_mud_mortar_stone']]\
.groupby('geo_level_1_id').transform('sum')
```

```python
df_test['avg_floors_per_year'] = df_test[['age', 'count_floors_pre_eq']].groupby('age').transform('mean')
```

```python
df_test['avg_area'] = df_test[['geo_level_1_id', 'area_percentage']].\
groupby('geo_level_1_id').transform('mean')
```

```python
encoder= ce.BinaryEncoder(cols= ['land_surface_condition', 'foundation_type', \
                            'roof_type', 'ground_floor_type', 'other_floor_type', 'position', 'plan_configuration', 'legal_ownership_status'], return_df = True)
```

#### Aplico binary Encoding a set de entrenamiento

```python
df_binary_encoding = encoder.fit_transform(train_values)
```

```python
df_binary_encoding
```

#### Aplico binary Encoding a set de test

```python
test_binary_encoding = encoder.fit_transform(df_test)
```

```python
test_binary_encoding
```

```python
test_binary_encoding.to_csv('/content/drive/MyDrive/datasets_tp2/test_binary.csv', index = False)
```

```python
df_binary_encoding.to_csv('/content/drive/MyDrive/datasets_tp2/train_binary.csv', index = False)
```
